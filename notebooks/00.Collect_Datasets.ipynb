```python
# I define the inicial dataset in the 'inic' variable (should be a string that contain the path of the file) 
# and in the 'dest' variable I do a list with the path of all the files I want concat

def HistMake (inic, dest):
    datatot = pd.read_csv(inic)
    datatot.columns= datatot.columns.str.lower()

    #I make a variable that would be mi fist dataframe
    for i in dest: 
        data0 = pd.read_csv(i)
        data0.columns= data0.columns.str.lower()

        datatot = pd.concat([data0,datatot], axis = 0).reset_index(drop=True)
    return datatot
        
def TwoMake (inic, dest):
    datatot = pd.read_csv(inic)
    datatot.columns= datatot.columns.str.lower()
    #I make a variable that would be mi fist dataframe
    data0 = pd.read_csv(dest)
    data0.columns= data0.columns.str.lower()
    datatot = pd.concat([data0,datatot], axis = 0).reset_index(drop=True)
    return datatot
```

*08.Residential_Area(ha)* - Surface for residential use

*09.Density(hab / ha)* - Density is the number of inhabitants per unit area

*10.Real_Density (room / ha)* - Net density is the number of inhabitants per unit of residential area

```python
density = ["../datasets/all/densitat_2015.csv", "../datasets/all/densitat_2016.csv", 
           "../datasets/all/densitat_2017.csv", "../datasets/all/densitat_2018.csv", "../datasets/all/densitat_2019.csv"]
```

```python
density_hist = HistMake("../datasets/all/densitat_2014.csv", density) 
density_hist 
```

```python
density_hist.columns = ['Year', 'Code_Dist', 'Name_Dist', 'Code_NBHD', 'Name_NBHD', 'Population', 
                        'Total_Area(ha)', 'Residential_Area(ha)', 'Density(hab/ha)', 'Real_Density(hab/ha)']
```

```python
density_hist.isnull().sum()
```

```python
#Check if Residential_Area(ha) == Population/Real_Density(hab/ha)
not_true = [ x for x in density_hist.index if round(density_hist.loc[x, "Population"] / density_hist.loc[x, "Real_Density(hab/ha)"],ndigits=1) != density_hist.loc[x, "Residential_Area(ha)"]]
len(not_true)
```

```python
for a in density_hist.index:
    if round(density_hist.loc[a, "Population"] / density_hist.loc[a, "Real_Density(hab/ha)"],ndigits=1) != density_hist.loc[a, "Residential_Area(ha)"]:
        print( round(density_hist.loc[a, "Population"] / density_hist.loc[a, "Real_Density(hab/ha)"],ndigits=1) , density_hist.loc[a, "Residential_Area(ha)"])
```

```python
#create a list to replace the null Values in the Residential_Area
density_hist["Residential_Area(ha)"].replace(np.nan, "null", inplace= True)
ccc = []
for x in density_hist.index:
    if density_hist.loc[x, "Residential_Area(ha)"] == 'null':
        ccc.append(round(density_hist.loc[x, "Population"] / density_hist.loc[x, "Real_Density(hab/ha)"],ndigits=1))
    else:
        ccc.append(density_hist.loc[x, "Residential_Area(ha)"])
```

```python
#Replace the Country with the list
density_hist["Residential_Area(ha)"] = ccc
```

```python
density_hist.to_csv(r'../datasets/Density_Hist.csv', index = False)
```

# Local by Use

RESOURCE STRUCTURE FROM 2018

*01.Year* - Year of the data

*02.Code_Dist* - District code

*03.Name_Dist* - Name of the district

*04.Code_NBHD* - Neighborhood code

*05.Name_NBHD* - Name of the neighborhood

*06.Censal_Section* - Census section

*07.Use* - Description of the main destination use

*08.Description* - Number, Area m2 and Value € of the cadastral premises

*09.Amount* - Number of premises, area and value per destination use


```python
local = TwoMake("../datasets/all/locals_us_desti_2018.csv", "../datasets/all/locals_us_desti_2019.csv")
local
```

```python
local.isnull().sum()
```

```python
local.columns = ['Year', 'Code_Dist', 'Name_Dist', 'Code_NBHD', 'Name_NBHD', 'Censal_Section' ,'Use', "Description" , 'Amount']
```

```python
#Replace the Catalan manual
local.Description.replace("Superfície_m2", "Superficie_m2", inplace= True)
```

```python
local
```

```python
local.Use.unique()
```

```python
nat_cat = local.Use.unique()
nat_eng = ['Housing', 'Parking', 'Commerce', 'Warehouse', 'Office','Hotel', 'Education', 'Health', 'Religious', 
           'Public Administration','Shows', 'Sports', 'Others', 'Industry','Unbuilt land', 'Not listed', 'Not listed']
zip_nat = zip(nat_cat, nat_eng)
dict_nat = dict(zip_nat)
local.Use.replace(dict_nat, inplace=True)
```

```python
local["KEY"] =  local["Year"].astype(str) +  "-" + local["Code_Dist"].astype(str) + "-" + local["Code_NBHD"].astype(str) + "-" + local["Censal_Section"].astype(str)+ "-" + local["Use"].astype(str)
```

```python
nom = local.loc[local.Description == "Nombre", :].reset_index(drop=True)
nom.columns = ['Year','Code_Dist','Name_Dist','Code_NBHD', 'Name_NBHD','Censal_Section' ,'Use', 'D','Amount', "Key"]
sup = local.loc[local.Description == "Superficie_m2", :].reset_index(drop=True)
sup.columns = ['Year','Code_Dist','Name_Dist','Code_NBHD', 'Name_NBHD','Censal_Section' ,'Use', 'D','Surface_M2', "Key"]
val = local.loc[local.Description == "Valor_€", :].reset_index(drop=True)
val.columns = ['Year', 'Code_Dist','Name_Dist','Code_NBHD', 'Name_NBHD','Censal_Section' ,'Use', 'D','Euros', "Key"]
```

```python
nom.drop(columns = 'D', inplace = True)
```

```python
sup = sup.iloc[:,-2:]
val = val.iloc[:,-2:]
```

```python
data = pd.merge(nom,sup, left_on="Key", right_on="Key", how='left')
data
```

```python
data = pd.merge(data,val, left_on="Key", right_on="Key", how='left')
data
```

```python
data.info()
```

```python
data.to_csv(r'../datasets/Type_Locals.csv', index = False)
```
