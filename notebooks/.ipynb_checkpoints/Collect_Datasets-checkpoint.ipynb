```python
# I define the inicial dataset in the 'inic' variable (should be a string that contain the path of the file) 
# and in the 'dest' variable I do a list with the path of all the files I want concat

def HistMake (inic, dest):
    datatot = pd.read_csv(inic)
    datatot.columns= datatot.columns.str.lower()

    #I make a variable that would be mi fist dataframe
    for i in dest: 
        data0 = pd.read_csv(i)
        data0.columns= data0.columns.str.lower()

        datatot = pd.concat([data0,datatot], axis = 0).reset_index(drop=True)
    return datatot
        
def TwoMake (inic, dest):
    datatot = pd.read_csv(inic)
    datatot.columns= datatot.columns.str.lower()
    #I make a variable that would be mi fist dataframe
    data0 = pd.read_csv(dest)
    data0.columns= data0.columns.str.lower()
    datatot = pd.concat([data0,datatot], axis = 0).reset_index(drop=True)
    return datatot
```

# Sales Price Historical

*01.Year* - Year of the data

*02.Trimester* - Data quarter

*03.Code_District* - District code

*04.Name_Dist* - Name of the district

*05.Code_NBHD* - Neighborhood code

*06.Name_NBHD* - Name of the neighborhood

*07.Description* - Average house price (thousands of euros)

*08.Price* - Value of the price of the house

*09.Type* -  The description of the type of the house


```python
price = [ "../datasets/all/comp_vend_preu_trim_2015.csv",
         "../datasets/all/comp_vend_preu_trim_2016.csv", "../datasets/all/comp_vend_preu_trim_2017.csv", 
         "../datasets/all/comp_vend_preu_trim_2018.csv", "../datasets/all/comp_vend_preu_trim_2019.csv"]
```

```python
price_hist = HistMake("../datasets/all/comp_vend_preu_trim_2014.csv", price) 
price_hist
```

```python
price_hist.columns = ['Year', 'Trimester', 'Code_District', 'Name_Dist', 'Code_NBHD', 'Name_NBHD', 'Description', 'Price']
price_hist
```

```python
price_hist["Type"] = price_hist.Description.apply(lambda x: x.split(".")[0])
price_hist["Description"] = price_hist.Description.apply(lambda x: x.split(".")[1])
```

```python
price_hist.isnull().sum()
```

```python
price_hist.Type.value_counts()
```

```python
#Replace the Catalan manual
price_hist.Description.replace(" Euros/m2 construït", "Euros/m2", inplace= True)
price_hist.Description.replace(" Milers d'euros", "Thousands of Euros", inplace= True)
price_hist.Type.replace("Usat", "Used", inplace= True)
price_hist.Type.replace("Nou", "New", inplace= True)
```

```python
price_hist
```

```python
price_hist.to_csv(r'../datasets/Sale_Houses_Hist.csv', index = False)
```
